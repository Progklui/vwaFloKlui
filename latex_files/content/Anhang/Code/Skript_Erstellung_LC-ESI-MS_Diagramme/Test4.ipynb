```python
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
import urllib
import numpy as np
import pandas as pd
url = "https://raw.githubusercontent.com/plotly/datasets/master/spectral.csv"
f = urllib.urlopen(url)
spectra=np.loadtxt(f, delimiter=',')

data = pd.io.parsers.read_csv("Kuerbis_Analyse_3h_LC-ESI-MS.csv")

helpvalue = ['617', '619', '629', '631', '633', '645', '647', '661', '675', '793', '795', '807', '821']

i = 0
verts = []
xs = data[['Zeit']].values

zs = [617, 619, 629, 631, 633, 645, 647, 661, 675, 793, 795, 807, 821]


traces = []
print(xs)
for h in helpvalue:
    ys = data[[helpvalue[i]]].values
    ys = np.hstack(ys)
    verts.append(list(zip(xs, ys)))
    i = i+1
    print(ys)
    traces.append(dict(
        z=ys,
        x=xs,
        y=zs,
        
        showscale=False,
        type='surface',
    ))

fig = { 'data':traces, 'layout':{'title':'Ribbon Plot'} }
iplot(fig, filename='ribbon-plot-python')
```
